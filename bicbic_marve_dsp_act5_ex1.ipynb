```python
import numpy as np
import matplotlib.pyplot as plt
from thinkdsp import TriangleSignal, Wave

# Define the parameters
frequency = 440  # Frequency in Hz
duration = 0.01  # Duration in seconds
amplitude = 1.0  # Amplitude of the signal

# Create a triangle signal
triangle_signal = TriangleSignal(freq=frequency, amp=amplitude)

# Create a time array
t = np.linspace(0, duration, int(44100 * duration), endpoint=False)  # Assuming a sample rate of 44100 Hz

# Generate the waveform
wave = triangle_signal.make_wave(duration=duration, start=0, framerate=44100)

# Plot the waveform
plt.figure(figsize=(10, 4))
plt.title("Triangle Signal Waveform")
wave.plot()
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from thinkdsp import TriangleSignal, Wave, Spectrum

# Define the parameters
frequency = 440  # Frequency in Hz
duration = 0.01  # Duration in seconds
amplitude = 1.0  # Amplitude of the signal

# Create a triangle signal
triangle_signal = TriangleSignal(freq=frequency, amp=amplitude)

# Create a time array
t = np.linspace(0, duration, int(44100 * duration), endpoint=False)  # Assuming a sample rate of 44100 Hz

# Generate the waveform
wave = triangle_signal.make_wave(duration=duration, start=0, framerate=44100)

# Create a Spectrum object from the waveform
spectrum = wave.make_spectrum()

# Print the amplitude and phase of the component at frequency 0
amplitude_at_0 = spectrum.hs[0]
phase_at_0 = np.angle(amplitude_at_0)

print(f"Amplitude at frequency 0: {abs(amplitude_at_0)}")
print(f"Phase at frequency 0: {phase_at_0} radians")

# Plot the waveform
plt.figure(figsize=(10, 4))
plt.title("Triangle Signal Waveform")
wave.plot()
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from thinkdsp import TriangleSignal, Wave, Spectrum

# Define the parameters
frequency = 440  # Frequency in Hz
duration = 0.01  # Duration in seconds
amplitude = 1.0  # Amplitude of the signal

# Create a triangle signal
triangle_signal = TriangleSignal(freq=frequency, amp=amplitude)

# Create a time array
t = np.linspace(0, duration, int(44100 * duration), endpoint=False)  # Assuming a sample rate of 44100 Hz

# Generate the waveform
wave = triangle_signal.make_wave(duration=duration, start=0, framerate=44100)

# Create a Spectrum object from the waveform
spectrum = wave.make_spectrum()

# Set spectrum.hs[0] to 100
spectrum.hs[0] = 100

# Make a new Wave from the modified Spectrum
new_wave = spectrum.make_wave()

# Plot the original and modified waveforms
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.title("Original Waveform")
wave.plot()
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid()

plt.subplot(1, 2, 2)
plt.title("Modified Waveform")
new_wave.plot()
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid()

plt.tight_layout()
plt.show()

```
