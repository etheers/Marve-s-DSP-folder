```python
# lets play the wave
wave = signal.make_wave(duration=1, framerate=20000)
wave.apodize()
wave.make_audio()
```

```python
# the spectrum and plot it up
spectrum = wave.make_spectrum()
spectrum.plot()
decorate(xlabel='Frequency (Hz)')
```
